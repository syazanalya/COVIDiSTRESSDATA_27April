# Penanganan Missing Values
Akan dilakukan imputasi mean/median pada variabel PSS10_avg karena persentasenya yang tidak terlalu tinggi

```python
#Dicari mean dari PSS10_avg
m_PSS10_avg= asiatimur["PSS10_avg"].mean()
m_PSS10_avg
```

```python
asiatimur["PSS10_avg"]=asiatimur["PSS10_avg"].fillna(m_PSS10_avg)
```

```python
asiatimur2["Dem_employment"].replace({1: "Full time employed",1:"Part time employed",1:"Self-employed",0: "Retired",0:"Student",0:"Not employed"}, inplace=True)
```
